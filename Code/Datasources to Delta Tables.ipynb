```
%sql
--- Query to create and optimize table IN_QueryAjax_Scaled

DROP TABLE IF EXISTS IN_QueryAjax_Scaled_Optimized;

CREATE TABLE IN_QueryAjax_Scaled_Optimized
USING delta
PARTITIONED BY (Player_id)

SELECT *
FROM IN_QueryAjax_Scaled;

OPTIMIZE IN_QueryAjax_Scaled_Optimized ZORDER BY (Recording_datetime);
```
